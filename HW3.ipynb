### <a id='toc3_4_1_'></a>[train.csv](#toc0_)


- row_id: (int64) ID code for the row.
- timestamp: (int64) the time in milliseconds between this user interaction and the first event completion from that user.
- user_id: (int32) ID code for the user.
- content_id: (int16) ID code for the user interaction
- content_type_id: (int8) 0 if the event was a question being posed to the user, 1 if the event was the user watching a lecture.
- task_container_id: (int16) Id code for the batch of questions or lectures. For example, a user might see three questions in a row before seeing the explanations for any of them. Those three would all share a task_container_id.
- user_answer: (int8) the user's answer to the question, if any. Read -1 as null, for lectures.
- answered_correctly: (int8) if the user responded correctly. Read -1 as null, for lectures.
- prior_question_elapsed_time: (float32) The average time in milliseconds it took a user to answer each question in the previous question bundle, ignoring any lectures in between. Is null for a user's first question bundle or lecture. Note that the time is the average time a user took to solve each question in the previous bundle.
- prior_question_had_explanation: (bool) Whether or not the user saw an explanation and the correct response(s) after answering the previous question bundle, ignoring any lectures in between. The value is shared across a single question bundle, and is null for a user's first question bundle or lecture. Typically the first several questions a user sees were part of an onboarding diagnostic test where they did not get any feedback.

metadata for the questions posed to users.


- question_id: foreign key for the train/test content_id column, when the content type is question (0).

- correct_answer: the answer to the question. Can be compared with the train user_answer column to check if the user was right.
- part: the relevant section of the TOEIC test.
- tags: one or more detailed tag codes for the question. The meaning of the tags will not be provided, but these codes are sufficient for clustering the questions together.


### <a id='toc3_4_3_'></a>[lectures.csv](#toc0_)


metadata for the lectures watched by users as they progress in their education.


- lecture_id: foreign key for the train/test content_id column, when the content type is lecture (1).
- part: top level category code for the lecture.
- tag: one tag codes for the lecture. The meaning of the tags will not be provided, but these codes are sufficient for clustering the lectures \* together.
- type_of: brief description of the core purpose of the lecture.

### <a id='toc5_1_1_'></a>[analyze_missing_values](#toc0_)


Подсчет пропущенных значений по каждому столбцу

```python
def analyze_missing_values(df):
    # Подсчет пропущенных значений по каждому столбцу
    missing_count = df.isnull().sum()

    # Подсчет процента пропущенных значений
    missing_percentage = round((missing_count / len(df)) * 100, 2)

    # Создание нового датафрейма
    missing_df = pd.DataFrame(
        {
            "Столбцы": missing_count.index,
            "Количество пропущенных строк": missing_count.values,
            "% пропущенных значений": missing_percentage.values,
        }
    )

    # Фильтрация только тех столбцов, где есть пропуски
    missing_df = missing_df[missing_df["Количество пропущенных строк"] > 0].reset_index(
        drop=True
    )

    return missing_df
```

### <a id='toc5_1_2_'></a>[compare_dataframes](#toc0_)


Сравнение двух таблиц на идентичность

```python
def compare_dataframes(df1, df2):
    # Проверка на совпадение формата (колонки, их типы)
    if df1.shape != df2.shape:
        return False, "DataFrames имеют разные размеры"

    if list(df1.columns) != list(df2.columns):
        return False, "DataFrames имеют разные колонки"

    # Проверка значений на соответствие
    for col in df1.columns:
        if not df1[col].equals(df2[col]):
            return False, f"Значения в колонке {col} отличаются"

    return True, "DataFrames идентичны"
```

### <a id='toc5_1_3_'></a>[optimize_dataframe](#toc0_)


Минимизирует использование памяти, изменяя числовые колонки на наименьшие подходящие типы данных

```python
def optimize_dataframe(df):

    for col in df.columns:
        col_type = df[col].dtype

        if pd.api.types.is_numeric_dtype(col_type):
            c_min = df[col].min()
            c_max = df[col].max()

            # для INT
            if pd.api.types.is_integer_dtype(col_type):
                if c_min >= np.iinfo(np.int8).min and c_max <= np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif (
                    c_min >= np.iinfo(np.int16).min and c_max <= np.iinfo(np.int16).max
                ):
                    df[col] = df[col].astype(np.int16)
                elif (
                    c_min >= np.iinfo(np.int32).min and c_max <= np.iinfo(np.int32).max
                ):
                    df[col] = df[col].astype(np.int32)
                else:
                    df[col] = df[col].astype(np.int64)

            # Для FLOAT
            elif pd.api.types.is_float_dtype(col_type):

                # пробуем float в int
                if (df[col] % 1 == 0).all():
                    if (
                        c_min >= np.iinfo(np.int8).min
                        and c_max <= np.iinfo(np.int8).max
                    ):
                        df[col] = df[col].astype(np.int8)
                    elif (
                        c_min >= np.iinfo(np.int16).min
                        and c_max <= np.iinfo(np.int16).max
                    ):
                        df[col] = df[col].astype(np.int16)
                    elif (
                        c_min >= np.iinfo(np.int32).min
                        and c_max <= np.iinfo(np.int32).max
                    ):
                        df[col] = df[col].astype(np.int32)
                    else:
                        df[col] = df[col].astype(np.int64)
            else:
                # сдаемся
                df[col] = df[col].astype(np.float32)

    return df
```

### <a id='toc5_1_4_'></a>[display_full_analysis](#toc0_)


info, describe и много чего еще.

Зависимость от **analyze_missing_values**

```python
def display_full_analysis(df, calculate_corr=False, rows_to_display=5):

    print(
        "_______________________________________________________________________________________"
    )
    print(
        "_________________________________________START_________________________________________"
    )

    print("_________________________________")

    print(f"\nhead")
    print(f"Первые {rows_to_display} строк DataFrame:\n")
    display(df.head(rows_to_display))

    print("_________________________________")

    print(f"\ntail")
    print(f"Последние {rows_to_display} строк DataFrame:\n")
    display(df.tail(rows_to_display))

    print("_________________________________")

    print(f"\ninfo")
    print(f"Основная информация:\n")
    display(df.info())

    print("_________________________________")

    print(f"\ndescribe")
    print(f"Описание числовых колонок:\n")

    # describe обычный, с форматированием
    result = df.describe()
    result_perc = result.copy()  # в процентах
    for col in result.columns:
        if pd.api.types.is_numeric_dtype(result[col]):
            if pd.api.types.is_float_dtype(result[col]):
                result[col] = result[col].map("{:.2f}".format)
            else:
                result[col] = result[col].map("{:,.0f}".format)
    display(result)

    # describe в процентах, с форматированием
    for col in result_perc.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            # % от общего
            result_perc.at["count", col] = (
                result_perc.at["count", col] / len(df)
            ) * 100

            # % от макс
            for stat in ["mean", "std", "min", "25%", "50%", "75%", "max"]:
                result_perc.at[stat, col] = (
                    result_perc.at[stat, col] / result_perc.at["max", col]
                ) * 100

    print(f"\nв процентах от максимального или количества(count)\n")
    display(result_perc)

    print("_________________________________")

    print(f"\ndescribe object")
    print(f"Описание категориальных колонок:\n")
    if df.select_dtypes(include=["object"]).shape[1] > 0:
        display(df.describe(include=["object"]))
    else:
        print("Нет столбцов типа 'object' ")

    print("_________________________________")

    print(f"\nnunique")
    print(f"Количество уникальных значений в каждой колонке:\n")
    display(pd.DataFrame(df.nunique(), columns=["Unique"]))

    print("_________________________________")

    print(f"\nvalue_counts")
    print(f"Распределение уникальных значений для категориальных колонок:\n")
    for col in df.select_dtypes(include=["object"]).columns:
        counts = df[col].value_counts()
        percentages = round(df[col].value_counts(normalize=True) * 100, 4)
        result = pd.DataFrame({"Количество": counts, "Процент": percentages})
        display(result)

    print("_________________________________")

    print(f"\nisnull sum")
    print(f"Общий анализ пропущенных значений:\n")
    display(pd.DataFrame(df.isnull().sum(), columns=["Missing"]))

    print("_________________________________")

    print(f"\nanalyze_missing_values")
    print(f"Подробный анализ пропущенных значений:\n")
    missing_df = analyze_missing_values(df)
    display(missing_df)

    print("_________________________________")

    print(f"\ncorr")
    if calculate_corr:
        print(f"Корреляционная матрица:\n")
        display(df.corr())
    else:
        print(f"\nКорреляционная матрица отключена (calculate_corr=False).")

    print("_________________________________")
    print(
        "__________________________________________END__________________________________________"
    )
    print(
        "_______________________________________________________________________________________"
    )
```

### <a id='toc5_1_5_'></a>[unique_values_summary](#toc0_)

Выводит статистику по колонке

```python
def unique_values_summary(df, col):

    result = []
    total_rows = len(df)
    total_rows = len(df)
    total_non_null = df[col].notna().sum()
    total_null = df[col].isna().sum()
    unique_values = df[col].value_counts()

    # Проверяем, если у нас всего 10 уникальных, то все выводим, или же первые и последние 3
    if len(unique_values) <= 10:
        for val, count in unique_values.items():
            result.append([val, count, (count / total_rows) * 100])
    else:
        first_3 = unique_values.head(3)
        last_3 = unique_values.tail(3)

        for val, count in first_3.items():
            result.append([val, count, (count / total_rows) * 100])

        # Разделитель
        result.append(["...", "...", "..."])

        for val, count in last_3.items():
            result.append([val, count, (count / total_rows) * 100])

    # посмотрим на пустые значения
    if total_null > 0:
        result.append(["<NaN>", total_null, round((total_null / total_rows) * 100, 2)])

    # Добавим Total От всех и уникальных, тоже интересно посмотртеь
    result.append(
        [
            "Total",
            total_rows,
            round((total_rows / total_rows) * 100, 2),
        ]
    )
    result = pd.DataFrame(result, columns=[col, "Count", "% of Total"])

    # форматируем, т.к. некоторые большие числа через e отображаются
    result["Count"] = result["Count"].apply(
        lambda x: f"{int(x):,}" if isinstance(x, (int, float)) else x
    )

    return result
```

#### <a id='toc6_4_1_3_'></a>[tags_seen_n](#toc0_)

Добавим колонку, где проверим видел ли уже user_id лекцию на тему в tags_lectures. Если они видел две tag из трех tags, то будет 2. А в tags_lectures_n, как ранее видели, будет 3.

```python
df = df.sort_values(by=["user_id", "timestamp"])
```

```python
column_name = "tags_seen_n"

df[column_name] = 0
tags_seen = set()
user_id_prev = None

for row in df.itertuples():

    user_id = row.user_id
    content_type_id = row.content_type_id

    if user_id != user_id_prev:
        tags_seen = set()
        user_id_prev = user_id

    # лекция
    if content_type_id == 1:

        # добавляем tag Лекции в set
        tag = int(row.tag)
        df.at[row.Index, column_name] = 1
        if tag != -1:
            tags_seen.add(tag)

    # вопрос
    elif content_type_id == 0:
        if row.tags != -1:
            tags = set(map(int, row.tags.split()))
            intersection_count = len(tags.intersection(tags_seen))
            df.at[row.Index, column_name] = intersection_count
        else:
            df.at[row.Index, column_name] = 0
```

```python
column_name = "tags_lectures_seen_n"

df[column_name] = 0
tags_seen = set()
user_id_prev = None

for row in df.itertuples():

    user_id = row.user_id
    content_type_id = row.content_type_id

    if user_id != user_id_prev:
        tags_seen = set()
        user_id_prev = user_id

    # лекция
    if content_type_id == 1:

        # добавляем tag Лекции в set
        tag = int(row.tag)
        df.at[row.Index, column_name] = 1
        if tag != -1:
            tags_seen.add(tag)

    # вопрос
    elif content_type_id == 0:
        if row.tags_lectures != -1:
            tags = set(map(int, row.tags_lectures.split()))
            intersection_count = len(tags.intersection(tags_seen))
            df.at[row.Index, column_name] = intersection_count
        else:
            df.at[row.Index, column_name] = 0
```
